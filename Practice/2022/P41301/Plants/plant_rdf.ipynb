```python
# Classes
Ill_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Болезнь")
Climate_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Climate")
Leaf_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Лист")
Fruit_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#FruitPlant")
Soil_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Soil")
Flower_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#FlowerPlant")
Plant_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Plant")
Height_class = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Height")

#object properties
family_op = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#family")
height_op = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#heght_of_plant")
plant_property_op = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#plantProperty")
is_sick_op = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Болеет")
part_of_op=URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#partOf")
prefer_climate_op = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#preferClimate")
flowering_period_op =URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#floweringPeriod")

#data properties
plant_name = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#name")
plant_name_ru = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#nameRu")
species_dp = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Вид")
height_dp = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#heightDescription")
acidity_dp = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#acidity")
flower_color = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#flowerColor")
light = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#light")
max_temperature = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#maxTemperature")
min_temperature = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#minTemperature")
singular_temperature = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#singularTemperature")
max_height = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#maxHeight")
min_height = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#minHeight")
measurement_height = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#heightMeasurement")

#individuals
spring_ind = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Spring")
summer_ind = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Summer")
winter_ind = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Winter")
autumn_ind = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Autumn")


```

```python
def AddPlantIndividual(id, rnm):
    individual = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#Plant_"+id)
    g.add((individual, RDF.type, Plant_class))
    g.add((individual, plant_name, Literal(id)))
    g.add((individual, plant_name_ru, Literal(rnm)))

    return individual

def AddFlowerIndividual(id, ptf,cl):
    individual = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#FlowerPlant_"+id)
    g.add((individual, RDF.type, Flower_class))
    g.add((individual, part_of_op, ptf))
    g.add((individual, flower_color, Literal(cl))) 

    return individual

def AddClimateIndividual(id, lt,mxt,mnt,srt):
    individual = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#ClimateForPlant_"+id)
    g.add((individual, RDF.type, Climate_class))
    g.add((individual, light, Literal(lt)))
    g.add((individual, max_temperature, Literal(mxt)))
    g.add((individual, min_temperature, Literal(mnt)))
    g.add((individual, singular_temperature, Literal(srt)))

    return individual

def AddHeightIndividual(id, hdtn,mxh,mnh,msmh,pt):
    individual = URIRef("http://www.semanticweb.org/yuri/ontologies/2022/1/plants-2#HeightPlant_"+id)
    g.add((individual, RDF.type, Height_class))
    g.add((individual, height_dp, Literal(hdtn)))
    g.add((individual, max_height, Literal(mxh)))
    g.add((individual, min_height, Literal(mnh)))
    g.add((individual, measurement_height, Literal(msmh)))
    g.add((individual, height_op, pt))

    return individual    
```
